```python
import torch
import torch.nn as nn
 
inp = torch.tensor([0.1,0.2]).reshape(1,-1)
weight_1 = torch.tensor([
    [.3,.25],
    [.4, .35]
])

weight_2 = torch.tensor([
    [.45, .4],
    [.7, .6]
])

lay_1=nn.Linear(in_features=2, out_features=2, bias=False)
lay_1.weight = nn.Parameter(weight_1)

lay_2 = nn.Linear(in_features=2, out_features=2, bias=False)
lay_2.weight = nn.Parameter(weight_2)

```

```python
class MyNeuralNetwork(nn.Module):
    def __init__(self):
        super(MyNeuralNetwork, self).__init__()
        
        self.layer1 = nn.Sequential(
            nn.Linear(in_features=2, out_features=2, bias=False),
            nn.Sigmoid()
        )
        
        self.layer2 = nn.Sequential(
            nn.Linear(in_features=2, out_features=2, bias=False),
            nn.Sigmoid()
        )
    
    def forward(self, x):
        x = self.layer1(x)
        x = self.layer2(x)
        
        return x

model = MyNeuralNetwork().to("cpu")
print(model)
```
